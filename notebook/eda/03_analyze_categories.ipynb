# Exploratory Data Analysis

```python
from pandas import concat, read_parquet

from data import file
import missingno as msno
import seaborn as sns
import pandas as pd
```

```python
data_train = read_parquet(file.news_articles_raw_train)
data_test = read_parquet(file.news_articles_raw_test)
```

## Distribution: Categorical

The distributions in training and test set are equally unbalanced. There is roughy one third of articles in least widespread category compared to the most widespread one.

### Training Set

```python
sns.set(rc={'figure.figsize': (15, 7)})
sns.countplot(x="label", data=data_train, order=sorted(data_train.label.unique()))
```

### Test Set

```python
sns.set(rc={'figure.figsize': (15, 7)})
sns.countplot(x="label", data=data_test, order=sorted(data_train.label.unique()))
```
