# Split

This notebook does a stratified split on label and stores training data, test data and validation data in a proportion of 0.7/0.2/0.1.

```python
from pandas import concat, read_parquet
from data import file
```

```python
data = read_parquet(file.news_articles_cleaned)
data
```

```python
from sklearn.model_selection import train_test_split

train_ratio = 0.70
validation_ratio = 0.20
test_ratio = 0.10

data_train, data_test_val = train_test_split(data, test_size=1 - train_ratio, stratify=data[['label']])
```

```python
# test is now 10% of the initial data set
# validation is now 15% of the initial data set
data_test, data_val = train_test_split(data_test_val, test_size=test_ratio/(test_ratio + validation_ratio), stratify=data_test_val[['label']]) 
```

```python
print("train size:", data_train.shape, "categories:", len(data_train.label.unique()))
print("test size :", data_test.shape, "categories:", len(data_test.label.unique()))
print("val size  :", data_val.shape, "categories:", len(data_val.label.unique()))
```

```python
data_train.to_parquet(path=file.news_articles_cleaned_train)
data_test.to_parquet(path=file.news_articles_cleaned_test)
data_val.to_parquet(path=file.news_articles_cleaned_val)
```
